# Running the script to collect the channel embeddings 


```python
import requests
import pandas as pd
import numpy as np

# API endpoints
USER_PROFILING_URL = "http://127.0.0.1:8000/user-profiling/"
EMBEDDING_URL = "http://127.0.0.1:8000/embed/channel-embedding"

# List of channel IDs to process
channel_ids =new_channel_ids

# Data collectors
channel_rows = []
embedding_rows = []
for idx, channel_id in enumerate(channel_ids, 1):
    print(f"[{idx}/{len(channel_ids)}] Processing channel ID: {channel_id}")

# for channel_id in channel_ids:
    # 1. Call user profiling API
    resp = requests.post(USER_PROFILING_URL, json={"channel_id": channel_id})
    if resp.status_code != 200:
        print(f"Error fetching {channel_id}: {resp.text}")
        continue

    data = resp.json()

    # Extract channel metadata
    channel_data = {
        "channel_title": data.get("channel_title"),
        "subscriber_count": data.get("subscriber_count"),
        "total_videos": data.get("total_videos"),
    }

    recent_videos = data.get("recent_videos", [])

    # Prepare row for channel CSV
    row = {
        "channel_id": channel_id,
        "channel_data": str(channel_data),  # store as string/dict
        "most_recent_vid": str(recent_videos[0]) if len(recent_videos) > 0 else np.nan,
    }

    # Add up to 10 more videos
    for i in range(1, 11):
        if i < len(recent_videos):
            row[f"video_{i}"] = str(recent_videos[i])
        else:
            row[f"video_{i}"] = np.nan

    channel_rows.append(row)

    # 2. Prepare payload for embedding API (excluding recent_videos[0])
    embed_payload = data.copy()
    embed_payload["recent_videos"] = recent_videos[1:11]

    embed_resp = requests.post(EMBEDDING_URL, json=embed_payload)
    if embed_resp.status_code != 200:
        print(f"Error embedding {channel_id}: {embed_resp.text}")
        continue

    embedding_data = embed_resp.json()

    embedding_rows.append({
        "channel_id": channel_id,
        "embedding_response": embedding_data
    })

# Save channel details CSV
df_channels = pd.DataFrame(channel_rows)
df_channels.to_csv("channel_data.csv", index=False)

# Save embeddings CSV
df_embeddings = pd.DataFrame(embedding_rows)
df_embeddings.to_csv("embeddings.csv", index=False)

print("✅ Data saved: channel_data.csv and embeddings.csv")
```
